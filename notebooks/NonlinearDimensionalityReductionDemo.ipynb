```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from sklearn.datasets.samples_generator import make_swiss_roll
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

# Swiss roll data

```python
n = 800
X, color = make_swiss_roll(n_samples=n, random_state=123)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color)
plt.title('Swiss Roll in 3D');
```

```python
from sklearn.decomposition import PCA
X_spca = PCA(n_components=2).fit_transform(X)

plt.figure(figsize=(8,6))
plt.scatter(X_spca[:, 0], X_spca[:, 1], c=color)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
from sklearn.manifold import Isomap
X_after_isomap = Isomap(15, 2).fit_transform(X)

plt.figure(figsize=(8,6))
plt.scatter(X_after_isomap[:, 0], X_after_isomap[:, 1], c=color)
plt.show()
```

```python
from sklearn.manifold import LocallyLinearEmbedding
X_after_LLE = LocallyLinearEmbedding(15, 2).fit_transform(X)

plt.figure(figsize=(8,6))
plt.scatter(X_after_LLE[:, 0], X_after_LLE[:, 1], c=color)
plt.show()
```

## Digits data

```python
from sklearn import datasets
digits = datasets.load_digits()

X_digits = digits.data
plt.imshow(X_digits[np.random.randint(len(X_digits))].reshape(8,8), cmap="gray")
```

```python
np.bincount(digits.target)
```

```python
X_digits_emb = Isomap(30, 2).fit_transform(X_digits)
plt.figure(figsize=(8,6))
plt.scatter(X_digits_emb[:, 0], X_digits_emb[:, 1], c=digits.target)
plt.show()
```

```python
X_digits_emb = LocallyLinearEmbedding(30, 2).fit_transform(X_digits)
plt.figure(figsize=(8,6))
plt.scatter(X_digits_emb[:, 0], X_digits_emb[:, 1], c=digits.target)
plt.show()
```
